```python
# === 0) Imports & Setup ===
import os, copy, time
import numpy as np, pandas as pd
import scipy.io
from PIL import Image
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import torchvision.transforms as T
from torchvision.models import resnet34, ResNet34_Weights
from sklearn.model_selection import GroupKFold, train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix,
    mean_absolute_error, mean_squared_error, r2_score, roc_curve
)
import matplotlib.pyplot as plt

# reproducibility
seed = 42
np.random.seed(seed)
torch.manual_seed(seed)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"\nüñ•Ô∏è  Using device: {device}\n")

# === 1) Load .mat and Patient-Level Split ===
mat = scipy.io.loadmat('/kaggle/input/dataset-of-bmode-fatty-liver-ultrasound-images/dataset_liver_bmodes_steatosis_assessment_IJCARS.mat')
entries = mat['data'][0]

vols, labels, fats = [], [], []
for e in entries:
    vols.append(e['images'])               # (10, H, W)
    labels.append(int(e['class'].squeeze()))
    fats.append(float(e['fat'].squeeze()))

vols   = np.stack(vols, axis=0)           # (55, 10, H, W)
labels = np.array(labels)
fats   = np.array(fats)

# Hold-out 20% patients for test
pidx = np.arange(len(labels))
tv_idx, test_idx = train_test_split(pidx, test_size=0.2,
                                     stratify=labels, random_state=seed)
vols_tv, vols_test = vols[tv_idx], vols[test_idx]
lbl_tv, lbl_test   = labels[tv_idx], labels[test_idx]
fat_tv, fat_test   = fats[tv_idx], fats[test_idx]

print(f"üìä Patients: total={len(labels)}, train/val={len(tv_idx)}, test={len(test_idx)}")

# === 2) Flatten each 3D ‚Üí 2D slice with group IDs ===
def flatten(vols_arr, lbls, fts, pat_ids):
    imgs, labs, fatv, groups = [], [], [], []
    for pi, pid in enumerate(pat_ids):
        for sl in range(vols_arr.shape[1]):
            imgs.append(vols_arr[pi, sl])
            labs.append(lbls[pi])
            fatv.append(fts[pi])
            groups.append(pid)
    return (np.stack(imgs), np.array(labs), np.array(fatv), np.array(groups))

X_tv, y_tv, f_tv, g_tv = flatten(vols_tv, lbl_tv, fat_tv, tv_idx)
X_te, y_te, f_te, g_te = flatten(vols_test, lbl_test, fat_test, test_idx)
print(f"üî¢ Flattened: train/val slices={len(X_tv)}, test slices={len(X_te)}\n")

# === 3) Dataset Class ===
class SliceDataset(Dataset):
    def __init__(self, X, y, fats, scaler, transform, train=True):
        self.X, self.y = X, y
        self.fats = fats
        self.transform = transform
        self.train = train
        # Fit or apply scaler on fats
        if train:
            self.fats_s = scaler.fit_transform(fats.reshape(-1,1)).flatten()
        else:
            self.fats_s = scaler.transform(fats.reshape(-1,1)).flatten()
    def __len__(self):
        return len(self.y)
    def __getitem__(self, idx):
        arr = self.X[idx].astype(np.float32)
        arr = (arr - arr.min())/(arr.ptp()+1e-8)*255
        img = Image.fromarray(arr.astype(np.uint8), 'L')
        img = self.transform(img)
        return img, self.y[idx], self.fats_s[idx], self.fats[idx]

# === 4) Model & Focal Loss ===
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, alpha=None):
        super().__init__(); self.gamma, self.alpha = gamma, alpha
    def forward(self, logits, targets):
        ce = F.cross_entropy(logits, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce)
        return ((1-pt)**self.gamma * ce).mean()

class HybridNet(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        self.backbone = resnet34(weights=ResNet34_Weights.IMAGENET1K_V1)
        in_f = self.backbone.fc.in_features
        self.backbone.fc = nn.Identity()
        self.cls_head = nn.Sequential(
            nn.Linear(in_f,256), nn.ReLU(), nn.Dropout(0.4),
            nn.Linear(256,num_classes)
        )
        self.reg_head = nn.Sequential(
            nn.Linear(in_f,256), nn.SiLU(), nn.Dropout(0.4),
            nn.Linear(256,1)
        )
    def forward(self, x):
        feats = self.backbone(x)
        return self.cls_head(feats), self.reg_head(feats).squeeze(1)

# === 5) Plot Helpers ===
def plot_cm(cm, title):
    plt.figure(figsize=(4,4))
    plt.imshow(cm, cmap='Blues')
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j,i,cm[i,j],ha='center',va='center',
                     color='white' if cm[i,j]>cm.max()/2 else 'black')
    plt.title(title); plt.xlabel('Pred'); plt.ylabel('True')
    plt.tight_layout(); plt.savefig(f"{title.replace(' ','_')}.png"); plt.show()

def plot_reg(y_true, y_pred, title):
    plt.figure(figsize=(5,5))
    plt.scatter(y_true, y_pred, alpha=0.6)
    m,n = min(y_true.min(),y_pred.min()), max(y_true.max(),y_pred.max())
    plt.plot([m,n],[m,n],'r--'); plt.title(title)
    plt.xlabel('True Fat%'); plt.ylabel('Pred Fat%')
    plt.tight_layout(); plt.savefig(f"{title.replace(' ','_')}.png"); plt.show()

# === 6) CV Training with GroupKFold ===
gkf = GroupKFold(n_splits=5)
best_f1, best_state, best_fold = -1, None, -1

# Transforms
tf_train = T.Compose([
    T.RandomHorizontalFlip(), T.RandomRotation(10),
    T.ToTensor(), T.Lambda(lambda t: t.repeat(3,1,1)),
    T.Normalize([0.5]*3,[0.5]*3)
])
tf_val = T.Compose([
    T.ToTensor(), T.Lambda(lambda t: t.repeat(3,1,1)),
    T.Normalize([0.5]*3,[0.5]*3)
])

for fold, (tr_idx, vl_idx) in enumerate(gkf.split(X_tv, y_tv, groups=g_tv), 1):
    print(f"\nüé≤ === Fold {fold} === üé≤")
    X_tr, X_vl = X_tv[tr_idx], X_tv[vl_idx]
    y_tr, y_vl = y_tv[tr_idx], y_tv[vl_idx]
    f_tr, f_vl = f_tv[tr_idx], f_tv[vl_idx]
    # sampler
    cnt = np.bincount(y_tr)
    sw  = (1.0/cnt)[y_tr]
    sampler = WeightedRandomSampler(sw, len(sw), True)
    alpha = torch.tensor((len(y_tr)/(2*cnt)).astype(np.float32), device=device)
    # scaler
    scaler = MinMaxScaler()
    # datasets & loaders
    ds_tr = SliceDataset(X_tr,y_tr,f_tr,scaler,tf_train, train=True)
    ds_vl = SliceDataset(X_vl,y_vl,f_vl,scaler,tf_val,   train=False)
    loader_tr = DataLoader(ds_tr, batch_size=8, sampler=sampler, num_workers=2)
    loader_vl = DataLoader(ds_vl, batch_size=8, shuffle=False, num_workers=2)
    # model, losses, optimizer, sched, earlystop
    model = HybridNet().to(device)
    # init bias = log priors
    pri = cnt/cnt.sum()
    model.cls_head[-1].bias.data = torch.log(torch.tensor(pri+1e-8, dtype=torch.float32)).to(device)
    crit_cls = FocalLoss(gamma=2.0, alpha=alpha)
    crit_reg = nn.L1Loss()
    opt      = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-4)
    sched    = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode='min', patience=3, verbose=False)
    class EarlyStopper:
        def __init__(self, patience=10, min_delta=1e-4):
            self.patience, self.min_delta = patience, min_delta
            self.best, self.counter = np.inf, 0; self.stop=False
        def __call__(self, loss):
            if loss < self.best - self.min_delta:
                self.best, self.counter = loss, 0
            else:
                self.counter+=1
                if self.counter>=self.patience: self.stop=True
    early = EarlyStopper(patience=10)

    # Per‚Äëepoch history
    history = []
    for ep in range(1,51):
        # ‚îÄ‚îÄ TRAIN ‚îÄ‚îÄ
        model.train()
        t_preds, t_labels = [], []
        t_reg_preds, t_reg_true = [], []
        t_loss_sum = 0
        for imgs, lbs, fs_s, fs in loader_tr:
            imgs, lbs, fs_s = imgs.to(device), lbs.to(device), fs_s.to(device)
            opt.zero_grad()
            logits, preds = model(imgs)
            loss = crit_cls(logits, lbs) + crit_reg(preds, fs_s)
            loss.backward(); opt.step()
            t_loss_sum += loss.item()*imgs.size(0)
            # accumulate
            t_preds.extend(logits.argmax(1).cpu().numpy())
            t_labels.extend(lbs.cpu().numpy())
            t_reg_preds.extend(preds.detach().cpu().numpy())
            t_reg_true.extend(fs_s.cpu().numpy())
        # Train metrics
        t_loss = t_loss_sum / len(ds_tr)
        # classification
        t_acc       = accuracy_score(t_labels, t_preds)
        t_prec      = precision_score(t_labels, t_preds, zero_division=0, average='macro')
        t_rec       = recall_score(   t_labels, t_preds, zero_division=0, average='macro')
        t_f1_macro  = f1_score(       t_labels, t_preds, zero_division=0, average='macro')
        t_f1_w      = f1_score(       t_labels, t_preds, zero_division=0, average='weighted')
        # for binary F1, if two classes:
        t_f1_bin    = f1_score(       t_labels, t_preds, zero_division=0, average='binary')
        # regression‚Äîdescale
        t_reg_true_orig = scaler.inverse_transform(np.array(t_reg_true).reshape(-1,1)).flatten()
        t_reg_pred_orig = scaler.inverse_transform(np.array(t_reg_preds).reshape(-1,1)).flatten()
        t_mae       = mean_absolute_error(t_reg_true_orig, t_reg_pred_orig)
        t_mse       = mean_squared_error(      t_reg_true_orig, t_reg_pred_orig)
        t_rmse      = np.sqrt(t_mse)
        t_mape      = np.mean(np.abs((t_reg_true_orig - t_reg_pred_orig)/ (t_reg_true_orig+1e-8))) * 100

        # ‚îÄ‚îÄ VALIDATION ‚îÄ‚îÄ
        model.eval()
        v_preds, v_labels = [], []
        v_reg_preds, v_reg_true = [], []
        v_loss_sum = 0
        with torch.no_grad():
            for imgs, lbs, fs_s, fs in loader_vl:
                imgs, lbs, fs_s = imgs.to(device), lbs.to(device), fs_s.to(device)
                logits, preds = model(imgs)
                loss = crit_cls(logits, lbs) + crit_reg(preds, fs_s)
                v_loss_sum += loss.item()*imgs.size(0)
                v_preds.extend(logits.argmax(1).cpu().numpy())
                v_labels.extend(lbs.cpu().numpy())
                v_reg_preds.extend(preds.cpu().numpy())
                v_reg_true.extend(fs_s.cpu().numpy())
        v_loss = v_loss_sum / len(ds_vl)
        v_acc       = accuracy_score(v_labels, v_preds)
        v_prec      = precision_score(v_labels, v_preds, zero_division=0, average='macro')
        v_rec       = recall_score(   v_labels, v_preds, zero_division=0, average='macro')
        v_f1_macro  = f1_score(       v_labels, v_preds, zero_division=0, average='macro')
        v_f1_w      = f1_score(       v_labels, v_preds, zero_division=0, average='weighted')
        v_f1_bin    = f1_score(       v_labels, v_preds, zero_division=0, average='binary')
        v_reg_true_o = scaler.inverse_transform(np.array(v_reg_true).reshape(-1,1)).flatten()
        v_reg_pred_o = scaler.inverse_transform(np.array(v_reg_preds).reshape(-1,1)).flatten()
        v_mae       = mean_absolute_error(v_reg_true_o, v_reg_pred_o)
        v_mse       = mean_squared_error(      v_reg_true_o, v_reg_pred_o)
        v_rmse      = np.sqrt(v_mse)
        v_mape      = np.mean(np.abs((v_reg_true_o - v_reg_pred_o)/(v_reg_true_o+1e-8))) * 100

        # logging
        history.append({
            'epoch': ep,
            # train metrics
            'tr_loss': t_loss, 'tr_acc': t_acc, 'tr_prec': t_prec, 'tr_rec': t_rec,
            'tr_f1_macro': t_f1_macro, 'tr_f1_weighted': t_f1_w, 'tr_f1_binary': t_f1_bin,
            'tr_mae': t_mae, 'tr_rmse': t_rmse, 'tr_mse': t_mse, 'tr_mape': t_mape,
            # val metrics
            'vl_loss': v_loss, 'vl_acc': v_acc, 'vl_prec': v_prec, 'vl_rec': v_rec,
            'vl_f1_macro': v_f1_macro, 'vl_f1_weighted': v_f1_w, 'vl_f1_binary': v_f1_bin,
            'vl_mae': v_mae, 'vl_rmse': v_rmse, 'vl_mse': v_mse, 'vl_mape': v_mape,
            'lr': opt.param_groups[0]['lr']
        })

        # Fancy print
        print(f"‚≠ê Fold {fold} ‚Ä¢ Epoch {ep} ‚Ä¢ "
              f"TR:[L={t_loss:.3f} A={t_acc:.3f} F1={t_f1_macro:.3f} MAE={t_mae:.2f}] | "
              f"VL:[L={v_loss:.3f} A={v_acc:.3f} F1={v_f1_macro:.3f} MAE={v_mae:.2f}]")

        sched.step(v_loss)
        early(v_loss)
        if early.stop:
            print("‚èπÔ∏è  EarlyStopping triggered\n")
            break

    # save epoch history
    pd.DataFrame(history).to_csv(f'history_fold{fold}.csv', index=False)

    # fold-level confusion & regression
    cm_v = confusion_matrix(v_labels, v_preds)
    plot_cm(cm_v, f"Fold{fold}_Val_CM")
    plot_reg(v_reg_true_o, v_reg_pred_o, f"Fold{fold}_Val_Reg")

    # fold summary print & save
    # --- Fold summary printout ---
    print(f"\nüéâ Fold {fold} Summary üéâ")
    print(f" Train ‚Üí Loss: {t_loss:.3f} | Acc: {t_acc:.3f} | Prec: {t_prec:.3f} | Rec: {t_rec:.3f} | "
      f"F1Œº: {t_f1_macro:.3f} | F1w: {t_f1_w:.3f} | F1b: {t_f1_bin:.3f} | MAE: {t_mae:.2f} | RMSE: {t_rmse:.2f}")
    print(f" Val   ‚Üí Loss: {v_loss:.3f} | Acc: {v_acc:.3f} | Prec: {v_prec:.3f} | Rec: {v_rec:.3f} | "
      f"F1Œº: {v_f1_macro:.3f} | F1w: {v_f1_w:.3f} | F1b: {v_f1_bin:.3f} | MAE: {v_mae:.2f} | RMSE: {v_rmse:.2f}")
    print("-" * 80)

    pd.DataFrame([{
        'fold': fold,
        'val_acc': v_acc, 'val_f1_macro': v_f1_macro, 'val_f1_weighted': v_f1_w,
        'val_f1_binary': v_f1_bin, 'val_prec': v_prec, 'val_rec': v_rec,
        'val_mae': v_mae, 'val_rmse': v_rmse, 'val_mse': v_mse, 'val_mape': v_mape
    }]).to_csv(f'val_metrics_fold{fold}.csv', index=False)

    if v_f1_macro > best_f1:
        best_f1, best_state, best_fold = v_f1_macro, copy.deepcopy(model.state_dict()), fold

# === 7) Test Set Evaluation ===
print(f"\nüöÄ Best fold: {best_fold} with Val F1={best_f1:.3f}")
model.load_state_dict(best_state)

# Build test loader
scaler_test = MinMaxScaler().fit(f_tv.reshape(-1,1))
ds_test = SliceDataset(X_te, y_te, f_te, scaler_test, tf_val, train=False)
loader_test = DataLoader(ds_test, batch_size=8, shuffle=False, num_workers=2)

# Eval
model.eval()
tp, tl, trp, trt = [], [], [], []
with torch.no_grad():
    for imgs, lbs, fs_s, fs in loader_test:
        imgs = imgs.to(device)
        logits, preds = model(imgs)
        probs = torch.softmax(logits, dim=1)[:,1].cpu().numpy()
        cls_pred = (probs>=0.5).astype(int)
        tp.extend(cls_pred); tl.extend(lbs.numpy())
        pr_o = scaler_test.inverse_transform(preds.cpu().numpy().reshape(-1,1)).flatten()
        trp.extend(pr_o); trt.extend(fs.numpy())

# compute metrics
test_acc = accuracy_score(tl, tp)
test_prec = precision_score(tl, tp, zero_division=0, average='macro')
test_rec  = recall_score(   tl, tp, zero_division=0, average='macro')
test_f1m  = f1_score(       tl, tp, zero_division=0, average='macro')
test_f1w  = f1_score(       tl, tp, zero_division=0, average='weighted')
test_f1b  = f1_score(       tl, tp, zero_division=0, average='binary')
test_mae  = mean_absolute_error(trt, trp)
test_mse  = mean_squared_error(      trt, trp)
test_rmse = np.sqrt(test_mse)
test_mape = np.mean(np.abs((np.array(trt)-np.array(trp))/(np.array(trt)+1e-8))) * 100
cm_test   = confusion_matrix(tl, tp)

# Styled print
print("üåü=== TEST SET RESULTS ===üåü")
print(f"Acc   : {test_acc:.3f}    Prec: {test_prec:.3f}    Rec: {test_rec:.3f}")
print(f"F1-Œº   : {test_f1m:.3f}    F1-w: {test_f1w:.3f}    F1-b: {test_f1b:.3f}")
print(f"MAE   : {test_mae:.2f}    RMSE: {test_rmse:.2f}    MAPE: {test_mape:.2f}%\n")

# save test metrics & plots
pd.DataFrame([{
    'test_acc': test_acc, 'test_prec': test_prec, 'test_rec': test_rec,
    'test_f1_macro': test_f1m, 'test_f1_weighted': test_f1w, 'test_f1_binary': test_f1b,
    'test_mae': test_mae, 'test_rmse': test_rmse, 'test_mse': test_mse, 'test_mape': test_mape
}]).to_csv('test_metrics.csv', index=False)

plot_cm(cm_test, "Test_CM")
plot_reg(np.array(trt), np.array(trp), "Test_Reg")

# === 8) Inference Helper ===
```

```python



def inference(volume_3d, true_lbl, true_fat, model, scaler, transform):
    model.eval()
    sl = volume_3d.shape[0] // 2
    arr = volume_3d[sl].astype(np.float32)
    arr = (arr - arr.min())/(arr.ptp()+1e-8)*255
    img = Image.fromarray(arr.astype(np.uint8),'L')
    inp = transform(img).unsqueeze(0).to(device)
    with torch.no_grad():
        logits, pred_s = model(inp)
    pred_cls = logits.argmax(1).item()
    pred_f = scaler.inverse_transform([[pred_s.item()]])[0][0]
    print(f"\nüîç True cls={true_lbl}, Pred cls={pred_cls}")
    print(f"    True fat={true_fat:.1f}%, Pred fat={pred_f:.1f}%")
    plt.figure(figsize=(5,5)); plt.imshow(img, cmap='gray'); plt.axis('off')
    plt.title(f"T{true_lbl}|P{pred_cls}  Tf{true_fat:.1f}%|Pf{pred_f:.1f}%"); plt.show()

# Example:
sample_idx=5

inference(vols_test[sample_idx], lbl_test[sample_idx], fat_test[sample_idx], model, scaler_test, tf_val)

```
